## Problem 1:Smart Parking Lot Management System

```python
def parking_management_system(vehicle_logs):
    capacity = 10
    current = 0
    peak = 0
    for log in vehicle_logs:
        if log == 'IN':
            current += 1
            if current > peak:
                peak = current

        if log == 'OUT':
            current -= 1
    if current > capacity:
        Parking_status = 'Unavailable'
    else:
        Parking_status = 'Available'

    print('Currently Parked Vehicles:', current)
    print("Peak Parking Usage:", peak)
    print('Parking Status:', Parking_status)

parking_management_system(["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

## Problem 2:Online Food Delivery Time Estimator

```python
def online_delivery_time(distance, traffic_level, weather):
    base_time = distance * 5
    delay = 0

    # Convert inputs to lowercase for case-insensitive comparison
    traffic_level_lower = traffic_level.lower()
    weather_lower = weather.lower()

    if traffic_level_lower == 'high':
        delay += 15
    elif traffic_level_lower == "medium":
        delay += 8

    if weather_lower == 'rainy':
        delay += 10
    elif weather_lower == "stormy":
        delay += 20

    final_eta = base_time + delay

    print("Estimated Delivery Time:", final_eta, "minutes")

online_delivery_time(9, "High", "Rainy")
```

## Problem 3:Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)

seat_occupancy_analyzer(200, [1] * 150)
```

## Problem 4:Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", average, "%")
    print("Server Status:", status)

server_load_classifier([45, 60, 70, 85, 90])
```

## Problem 5:Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

## Problem 6:Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration_controller(100, 105)
```
