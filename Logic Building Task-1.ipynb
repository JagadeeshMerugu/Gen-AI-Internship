## 1. User Login Check

Problem Statement:

Given a username and password, check whether login is successful.

username = "admin"

password = "1234"

Requirements:

Print "Login Successful" if both username and password match

Otherwise print "Invalid Credentials"

Real-World Application: Authentication systems


```python
# Given credentials
correct_username = "admin"
correct_password = "1234"

username = input("Enter username: ")
password = input("Enter password: ")

#Check login credentials
if username == correct_username and password == correct_password:
    print("✓ Login Successful")
else:
    print("✗ Invalid Credentials")

print()
```

## 2. Pass / Fail Analyzer

Problem Statement:

Analyze student results.

marks = [45, 78, 90, 33, 60]

Requirements:

A student passes if marks ≥ 50

Count the total number of pass students

Count the total number of fail students

Print the final result clearly

Real-World Application: Academic evaluation systems


```python
#Given marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0
passing_marks = 50

# Analyze each student's marks
for i, mark in enumerate(marks, 1):
    if mark >= passing_marks:
        pass_count += 1
        status = "PASS"
    else:
        fail_count += 1
        status = "FAIL"
    print(f"Student {i}: {mark} marks - {status}")

# Print final results
print("\n--- ANALYSIS SUMMARY ---")
print(f"Total Students: {len(marks)}")
print(f"Passed Students: {pass_count}")
print(f"Failed Students: {fail_count}")
print(f"Pass Percentage: {(pass_count/len(marks))*100:.2f}%")

print()
```

## 3. Simple Data Cleaner

Problem Statement:

Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]

Requirements:

Remove extra spaces

Convert all names to lowercase

Print the cleaned list

Real-World Application: Data preprocessing before analysis


```python
names = [" Alice ", "bob", " CHARLIE "]

print("Original names:", names)

# Clean the data: remove spaces and convert to lowercase
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()  # strip() removes spaces, lower() converts to lowercase
    cleaned_names.append(cleaned_name)

print("Cleaned names:", cleaned_names)
```

## 4. Message Length Analyzer
Problem Statement:

Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:

Print the length of each message

Flag messages longer than 10 characters

Real-World Application: Text filtering and validation systems


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for i, message in enumerate(messages, 1):
    msg_length = len(message)
    
    if msg_length > 10:
        flag = "LONG MESSAGE"
    else:
        flag = "Normal"
    
    print(f"Message {i}: '{message}' | Length: {msg_length} | {flag}")
```

## 5. Error Message Detector

Problem Statement:

Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:

Count the number of "ERROR" entries

Print the total error count

Real-World Application: Monitoring and log analysis systems


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

print("System Logs:", logs)

# Count ERROR entries
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1


print(f"\n Total ERROR count: {error_count}")
```
