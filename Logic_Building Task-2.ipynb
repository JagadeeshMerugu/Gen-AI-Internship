### Problem 1: Unique Words in a Sentence

```python
def unique_words(sentence):

    words = sentence.lower().split()
    unique_words = set(words)

    print(f"Unique words count: {len(unique_words)}")
    print(f"Unique words: {unique_words}")
    
sentence = input('Enter Sentence: ')
unique_words(sentence)
```

### Problem 2: Highest Salary from Employee Data

```python
# Employee data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

### Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Method 1: Using built-in functions
max_value = max(numbers)
min_value = min(numbers)

# Display results
print(f"List: {numbers}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")

# Method 2: Manual approach (bonus)

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print(f"Max (manual): {maximum}")
print(f"Min (manual): {minimum}")
```

### Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

# Count products above 1000
threshold = 1000
count = 0

for price in prices:
    if price > threshold:
        count += 1
        
print(f"Products above {threshold}: {count}")
```

### Problem 5: Calculate Attendance Percentage

```python
# Method 1
attendance = ["P", "P", "A", "P", "P"]

#Count present days
present_count = attendance.count("P")
total_days = len(attendance)

# Calculate percentage
attendance_percentage = (present_count / total_days) * 100

# Display result
print(f"Attendance Percentage: {attendance_percentage}")

# Detailed breakdown
print(f"\nPresent: {present_count} days")
print(f"Absent: {attendance.count('A')} days")
print(f"Total: {total_days} days")
```

```python
# Method - 2

present_count = 0
total_days = len(attendance)

for ch in attendance:
    if ch.lower() == 'p':
        present_count += 1

# Calculate percentage
attendance_percentage = (present_count / total_days) * 100
        
# Display result
print(f"Attendance Percentage: {attendance_percentage}")

# Detailed breakdown
print(f"\nPresent: {present_count} days")
print(f"Absent: {attendance.count('A')} days")
print(f"Total: {total_days} days")
```

### Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

# Remove duplicates by converting to set
unique_phone_numbers = set(phone_numbers)

# Display result
print(f"Original phone numbers: {phone_numbers}")
print(f"Unique phone numbers: {unique_phone_numbers}")
print(f"Duplicates removed: {len(phone_numbers) - len(unique_phone_numbers)}")
```

### Problem 7: Count Character Frequency

```python
text = input('Enter a text: ')

char_frequency = {}

for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print(f"Character frequency: {char_frequency}")
```

```python
# Method -2
text = input('Enter a text: ')

char_freq_alt = {}
for char in text:
    char_freq_alt[char] = char_freq_alt.get(char, 0) + 1
print(f"Character frequency: {char_freq_alt}")
```

### Problem 9: Check if a Key Exists in a Dictionary

### Problem 10: Calculate Average Marks

```python
# Student marks
marks = [85, 90, 78, 62, 70]

# Calculate average
total_marks = sum(marks)
average_marks = total_marks / len(marks)

# Display result
print(f"Marks: {marks}")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks}")
```
