## Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")

#instance
light = SmartLight('Bedroom Light')
light.display_status()
light.turn_on() # action
light.display_status()
```

## Problem 2: Employee ID Card System

```python
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name       = name
        self.emp_id     = emp_id
        self.department = department

    def display_card(self):
        
        print("      Employee ID Card     ")
        print("---------------------------")
        print(f"  Name       : {self.name}")
        print(f"  ID         : {self.emp_id}")
        print(f"  Department : {self.department}")
        print("---------------------------")
        
employees = [
    ("Rahul Sharma",  "EMP102", "IT"),
    ("Priya Nair",    "EMP215", "Human Resources"),
    ("Arjun Mehta",   "EMP330", "Finance"),
]

for name, eid, dept in employees:
    # instance/ object creation
    emp = EmployeeIDCard(name, eid, dept) 
    emp.display_card()
```

## Problem 3: Mobile Contact Record

```python
class MobileContact:
    def __init__(self, contact_name, phone_number):
        self.contact_name = contact_name
        self.phone_number = phone_number

    def display_contact(self):
        print("--- Contact Saved ---")
        print(f"  Name  : {self.contact_name}")
        print(f"  Phone : {self.phone_number}")
        
        
# Object creation
contact = MobileContact("Jagadeesh", "9876543210")
contact.display_contact()
```

## Problem 4: Product Price Tag Generator

```python
class ProductPriceTag:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price        = price

    def display_tag(self):
        print("-----------------------------")
        print(f"  Product : {self.product_name}")
        print(f"  Price   : ₹{self.price:,}")
        print("-----------------------------")


products = [
    ("Headphones",         2499),
    ("Mechanical Keyboard", 4999),
    ("USB-C Hub",          1799),
]

for name, price in products:
    tag = ProductPriceTag(name, price)
    tag.display_tag()
```

## Problem 5: Movie Rating Display System

```python
class MovieRating:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating     = rating

    def display_rating(self):
        stars = "★" * int(self.rating) + "☆" * (5 - int(self.rating))
        print(f"  Movie  : {self.movie_name}")
        print(f"  Rating : {self.rating} / 5  {stars}")
        print()




movie = MovieRating("Pushpa 2", 5)
movie.display_rating()
```

## Problem 6: Delivery Address Manager

```python
class DeliveryAddress:
    def __init__(self, customer_name, address, city, pincode):
        self.customer_name = customer_name
        self.address       = address
        self.city          = city
        self.pincode       = pincode

    def display_delivery_details(self):
        print("============================")
        print("      Delivery Details      ")
        print("============================")
        print(f"  Customer : {self.customer_name}")
        print(f"  Address  : {self.address}")
        print(f"  City     : {self.city}")
        print(f"  Pincode  : {self.pincode}")
        print("============================")
        print()



deliveries = [
    ("Suman Rao",    "Plot 12, Jubilee Hills", "Hyderabad", "500033"),
    ("Kavya Menon",  "Flat 4B, Anna Nagar",    "Chennai",   "600040"),
    ("Nikhil Joshi", "203, MG Road",            "Bangalore", "560001"),
]

for name, addr, city, pin in deliveries:
    delivery = DeliveryAddress(name, addr, city, pin)
    delivery.display_delivery_details()
```
