**Basic LLM Call**

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

response = llm.invoke("Explain LangChain in simple terms")
print(response.content)
```

**Prompt Template**

```python
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms with example"
)

prompt = template.format(topic="Artificial Intelligence")
print(prompt)

# Run with LLM
response = llm.invoke(prompt)
print(response.content)
```

**Simple Chain**

```python
from langchain.chains import LLMChain

chain = LLMChain(
    llm=llm,
    prompt=template
)

output = chain.run("Machine Learning")
print(output)
```

**Memory**

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)

print(conversation.run("Hi, My name is Jagadeesh"))
print(conversation.run("What is my name?"))
```

**Create Custom Tool**

```python
from langchain.tools import Tool

def multiply_numbers(input_text):
    return str(eval(input_text))

tool = Tool(
    name="Calculator",
    func=multiply_numbers,
    description="Useful for solving math problems like 5*6"
)
```

**Agent with Tool**

```python
from langchain.tools import Tool
from langchain.agents import initialize_agent

def calculator(input_text):
    return str(eval(input_text))

tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Useful for solving math expressions like 5*6"
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description",
    verbose=True
)

print(agent.run("What is 45 * 10?"))
```

**Document Loader**

```python
from langchain.document_loaders import TextLoader

# Create sample file
with open("sample.txt", "w") as f:
    f.write("LangChain helps build LLM applications with memory and tools.")

loader = TextLoader("sample.txt")
documents = loader.load()

print(documents)
```

**Vector Store (FAISS)**

```python
from langchain.vectorstores import FAISS, faiss
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()

vectorstore = FAISS.from_documents(documents, embeddings)

query = "What is LangChain?"
docs = vectorstore.similarity_search(query)

print(docs[0].page_content)
```

**Simple RAG (Retrieval QA)**

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever()
)

response = qa_chain.run("What does LangChain do?")
print(response)
```

**Modular Clean Pipeline**

```python
def build_llm():
    return ChatGroq(model="llama-3.3-70b-versatile")

def build_prompt():
    return PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} in simple terms"
    )

def build_chain(llm, prompt):
    return LLMChain(llm=llm, prompt=prompt)

llm = build_llm()
prompt = build_prompt()
chain = build_chain(llm, prompt)

print(chain.run("Neural Networks"))
```
