##  Task 1: Replace hardcoded prompts with PromptTemplate

```python

class Task1PromptSystem:
    """Modular system for Task 1 - No hardcoding, no f-strings"""

    def __init__(self):
        # Template separated from logic
        self.template = PromptTemplate(
            input_variables=["topic"],  # ✅ Using input_variables
            template="Explain {topic} in simple terms for beginners"
        )

    def generate_prompt(self, topic):
        """Generate prompt - logic separate from template"""
        return self.template.format(topic=topic)

# Test the implementation
print("="*60)
print("TASK 1: Replace Hardcoded Prompts")
print("="*60)

task1 = Task1PromptSystem()
test_topics = ["Artificial Intelligence", "Python", "Machine Learning"]

for topic in test_topics:
    result = task1.generate_prompt(topic)
    print(f"\nInput: {topic}")
    print(f"Output: {result}")
    print("-" * 40)

```

**Task 2: Multi-input prompt system with validation-ready design**

```python
class MultiInputPromptSystem:
    """Handles multiple inputs: topic, audience, tone"""

    def __init__(self):
        # Template with 3 input variables
        self.template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],  # ✅ All inputs declared
            template="Explain {topic} for {audience} in a {tone} tone"
        )

    def generate_prompt(self, topic, audience, tone):
        """Generate prompt with all three inputs"""
        return self.template.format(
            topic=topic,
            audience=audience,
            tone=tone
        )

print("\n" + "="*60)
print("TASK 2: Multi-Input Prompt System")
print("="*60)

task2 = MultiInputPromptSystem()

# Test cases from assignment
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for test in test_cases:
    result = task2.generate_prompt(**test)
    print(f"\nInput: {test}")
    print(f"Output: {result}")
    print("-" * 40)

print("\n✅ All three input variables used!")
print("✅ Template reusable with any combination!")
```

**Task 3: Three different PromptTemplates for different styles**

```python
class PromptVariationsEngine:
    """Engine for creating different prompt variations"""

    def __init__(self):
        # Template 1: Teaching style
        self.teaching_template = PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} clearly step by step"
        )

        # Template 2: Interview style
        self.interview_template = PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 questions about {topic}"
        )

        # Template 3: Storytelling style
        self.storytelling_template = PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} as a story"
        )

    def generate_all_variations(self, topic):
        """Generate all three prompt variations for a topic"""
        return {
            "teaching": self.teaching_template.format(topic=topic),
            "interview": self.interview_template.format(topic=topic),
            "storytelling": self.storytelling_template.format(topic=topic)
        }

print("\n" + "="*60)
print("TASK 3: Prompt Variations Engine")
print("="*60)

task3 = PromptVariationsEngine()
topic = "Machine Learning"

print(f"\nTopic: {topic}\n")
variations = task3.generate_all_variations(topic)

for style, prompt in variations.items():
    print(f"{style.upper()} Style:")
    print(f"  {prompt}")
    print()

print("✅ Three separate templates created!")
print("✅ Each template uses input_variables correctly!")
```

**Task 4: Chat-based prompt system with roles**

```python
class ChatPromptSystem:
    """Chat system with system and user roles"""

    def __init__(self):
        # Role configurations stored separately
        self.role_configs = {
            "teacher": {
                "system": "You are an expert teacher who explains concepts clearly and patiently.",
                "user": "Explain {topic} in detail with examples."
            },
            "interviewer": {
                "system": "You are a technical interviewer assessing candidate knowledge.",
                "user": "Ask me challenging questions about {topic}."
            },
            "motivator": {
                "system": "You are an enthusiastic motivator who inspires learning.",
                "user": "Help me understand {topic} with encouraging words."
            }
        }

    def generate_chat_prompt(self, role, topic):
        """Generate chat prompt with system and user messages"""
        if role not in self.role_configs:
            raise ValueError(f"Invalid role. Choose from: {list(self.role_configs.keys())}")

        config = self.role_configs[role]

        # Create chat template dynamically
        chat_template = ChatPromptTemplate.from_messages([
            ("system", config["system"]),
            ("user", config["user"])
        ])

        # Format and return
        messages = chat_template.format_messages(topic=topic)
        return {
            "system_prompt": messages[0].content,
            "user_prompt": messages[1].content
        }

print("\n" + "="*60)
print("TASK 4: ChatPromptTemplate System")
print("="*60)

task4 = ChatPromptSystem()

test_scenarios = [
    ("teacher", "Neural Networks"),
    ("interviewer", "Python Programming"),
    ("motivator", "Learning AI")
]

for role, topic in test_scenarios:
    print(f"\nRole: {role.upper()}")
    print(f"Topic: {topic}")
    print("-" * 40)

    result = task4.generate_chat_prompt(role, topic)
    print(f"System: {result['system_prompt']}")
    print(f"User: {result['user_prompt']}")
    print()

print("✅ ChatPromptTemplate used correctly!")
print("✅ System and User roles separated!")
```

**Task 5: Input validation layer (logic separate from templates)**

```python
class InputValidator:
    """Validation layer - completely separate from prompt templates"""

    VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
    VALID_TONES = ["formal", "casual", "fun"]

    @classmethod
    def validate_inputs(cls, audience, tone, strict_mode=False):
        """
        Validate audience and tone inputs
        Logic completely separate from templates
        """
        # Validate audience
        if audience.lower() not in cls.VALID_AUDIENCES:
            if strict_mode:
                raise ValueError(f"Invalid audience. Must be one of: {cls.VALID_AUDIENCES}")
            else:
                print(f"⚠️  Warning: '{audience}' invalid → defaulting to 'beginner'")
                validated_audience = "beginner"
        else:
            validated_audience = audience.lower()

        # Validate tone
        if tone.lower() not in cls.VALID_TONES:
            if strict_mode:
                raise ValueError(f"Invalid tone. Must be one of: {cls.VALID_TONES}")
            else:
                print(f"⚠️  Warning: '{tone}' invalid → defaulting to 'formal'")
                validated_tone = "formal"
        else:
            validated_tone = tone.lower()

        return validated_audience, validated_tone

    @classmethod
    def get_valid_options(cls):
        """Return valid options for reference"""
        return {
            "audiences": cls.VALID_AUDIENCES,
            "tones": cls.VALID_TONES
        }

print("\n" + "="*60)
print("TASK 5: Input Validation Layer")
print("="*60)

# Test valid inputs
print("\n✅ Test 1: Valid inputs")
audience, tone = InputValidator.validate_inputs("expert", "formal")
print(f"   Result: audience={audience}, tone={tone}")

# Test invalid with auto-correction
print("\n🔄 Test 2: Invalid inputs (auto-correction)")
audience, tone = InputValidator.validate_inputs("kids", "technical")
print(f"   Result: audience={audience}, tone={tone}")

# Test strict mode
print("\n🔒 Test 3: Strict mode")
try:
    audience, tone = InputValidator.validate_inputs("invalid", "invalid", strict_mode=True)
except ValueError as e:
    print(f"   ✓ Error caught: {e}")

print("\n📋 Valid Options:")
options = InputValidator.get_valid_options()
print(f"   Audiences: {options['audiences']}")
print(f"   Tones: {options['tones']}")

print("\n✅ Validation logic completely separate from templates!")
```

**Task 6: Complete Prompt Generator App**

```python
class PromptGeneratorApp:
    """Complete prompt generator with validation and multiple styles"""

    def __init__(self):
        # Style templates - separated from logic
        self.style_templates = {
            "teaching": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Explain {topic} for {audience} level learners in a {tone} teaching style"
            ),
            "interview": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Create interview questions about {topic} for {audience} candidates with a {tone} approach"
            ),
            "storytelling": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Tell a story about {topic} for {audience} audience in a {tone} storytelling style"
            )
        }

        # Validator instance (logic separation)
        self.validator = InputValidator

    def generate_prompt(self, topic, audience, tone, style):
        """
        Generate prompt with validation
        - No hardcoding
        - All inputs dynamic
        - Logic separate from templates
        """
        # Step 1: Validate inputs (separate logic)
        validated_audience, validated_tone = self.validator.validate_inputs(audience, tone)

        # Step 2: Check style exists
        if style.lower() not in self.style_templates:
            raise ValueError(f"Invalid style. Choose from: {list(self.style_templates.keys())}")

        # Step 3: Generate prompt using template
        template = self.style_templates[style.lower()]
        prompt = template.format(
            topic=topic,
            audience=validated_audience,
            tone=validated_tone
        )

        # Step 4: Return comprehensive result
        return {
            "original_input": {
                "topic": topic,
                "audience": audience,
                "tone": tone,
                "style": style
            },
            "validated_input": {
                "audience": validated_audience,
                "tone": validated_tone
            },
            "generated_prompt": prompt,
            "style_used": style.lower()
        }

print("\n" + "="*60)
print("TASK 6: Prompt Generator App")
print("="*60)

task6 = PromptGeneratorApp()

# Test Case 1
print("\n📝 Test Case 1: Storytelling style")
result = task6.generate_prompt(
    topic="Neural Networks",
    audience="beginners",
    tone="fun",
    style="storytelling"
)

print(f"Original: {result['original_input']}")
print(f"Validated: {result['validated_input']}")
print(f"Generated: {result['generated_prompt']}")
print(f"Style: {result['style_used']}")

# Test Case 2
print("\n📝 Test Case 2: Teaching style with invalid inputs")
result2 = task6.generate_prompt(
    topic="Python",
    audience="kids",
    tone="friendly",
    style="teaching"
)

print(f"Original: {result2['original_input']}")
print(f"Generated: {result2['generated_prompt']}")
```

**Task 7: Template Reusability Test**

```python
class ReusabilityTest:
    """Test one template with multiple inputs"""

    def __init__(self):
        # SINGLE reusable template - defined once
        self.reusable_template = PromptTemplate(
            input_variables=["concept", "difficulty", "purpose"],
            template="Explain the concept of {concept} at {difficulty} difficulty level for the purpose of {purpose}"
        )

    def test_with_multiple_inputs(self, test_cases):
        """Run the same template with different inputs"""
        results = []
        for i, test_case in enumerate(test_cases, 1):
            prompt = self.reusable_template.format(**test_case)
            results.append({
                "test_number": i,
                "input": test_case,
                "output": prompt
            })
        return results

print("\n" + "="*60)
print("TASK 7: Template Reusability Test")
print("="*60)

task7 = ReusabilityTest()

# 5 different test cases - SAME template structure
test_cases = [
    {"concept": "Machine Learning", "difficulty": "beginner", "purpose": "learning basics"},
    {"concept": "Deep Learning", "difficulty": "intermediate", "purpose": "building models"},
    {"concept": "Neural Networks", "difficulty": "advanced", "purpose": "research"},
    {"concept": "Natural Language Processing", "difficulty": "beginner", "purpose": "text analysis"},
    {"concept": "Computer Vision", "difficulty": "intermediate", "purpose": "image recognition"}
]

print(f"\n📋 Reusable Template Structure:")
print(f"   {task7.reusable_template.template}\n")

print("🎯 Running 5 different inputs through SAME template:\n")
results = task7.test_with_multiple_inputs(test_cases)

for result in results:
    print(f"Test {result['test_number']}:")
    print(f"  Input: {result['input']}")
    print(f"  Output: {result['output']}")
    print()

print("✅ Same template structure, 5 different outputs!")
print("✅ Perfect demonstration of reusability!")
```

**Complete Pipeline: User Input → Validation → Template → Output**

```python
class CompletePipeline:
    """Demonstrates the entire flow from input to output"""

    def __init__(self):
        self.validator = InputValidator
        self.generator = PromptGeneratorApp()

    def run(self, topic, audience, tone, style):
        """Run complete pipeline"""
        print("\n" + "="*60)
        print("COMPLETE PIPELINE FLOW")
        print("="*60)

        # Step 1: User Input
        print("\n1️⃣ USER INPUT (No hardcoding)")
        print(f"   → Topic: {topic}")
        print(f"   → Audience: {audience}")
        print(f"   → Tone: {tone}")
        print(f"   → Style: {style}")

        # Step 2: Validation Layer
        print("\n2️⃣ VALIDATION LAYER (Logic separate)")
        validated_audience, validated_tone = self.validator.validate_inputs(audience, tone)
        print(f"   → Validated Audience: {validated_audience}")
        print(f"   → Validated Tone: {validated_tone}")

        # Step 3: Prompt Template Selection
        print("\n3️⃣ PROMPT TEMPLATE (Using input_variables)")
        template = self.generator.style_templates.get(style.lower())
        if template:
            print(f"   → Template: {template.template}")

        # Step 4: Dynamic Generation
        print("\n4️⃣ DYNAMIC PROMPT GENERATION")
        result = self.generator.generate_prompt(topic, audience, tone, style)
        print(f"   → Generated: {result['generated_prompt']}")

        # Step 5: Final Output
        print("\n5️⃣ FINAL OUTPUT")
        print("   " + "="*50)
        print(f"   {result['generated_prompt']}")
        print("   " + "="*50)

        return result

# Run the complete pipeline
pipeline = CompletePipeline()

final_result = pipeline.run(
    topic="Generative AI",
    audience="beginner",
    tone="friendly",
    style="storytelling"
)

print("\n" + "="*60)
print("🎉 PIPELINE EXECUTION COMPLETE!")
print("="*60)
```
