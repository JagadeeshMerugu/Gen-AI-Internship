### Problem 1: Employee Performance Bonus Eligibility

```python
def find_top_performers(employees):
    
    #Find employees with the highest performance score.

    
    # Find the maximum score
    max_score = max(employees.values())
    
    # Find all employees with the maximum score
    top_performers = [name for name, score in employees.items() if score == max_score]
    
    # Format and display the result
    performers_str = ", ".join(top_performers)
    print(f"Top Performers Eligible for Bonus: {performers_str} (Score: {max_score})")
    
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

print(f"Input: {employees}")
# Calling the function
find_top_performers(employees)
```

### Problem 2: Search Query Keyword Analysis

```python
def analyze_search_keywords(query):
    
    # Convert to lowercase and split into words
    keywords = query.lower().split()
    
    # Count frequency of each keyword
    keyword_count = {}
    for keyword in keywords:
        keyword_count[keyword] = keyword_count.get(keyword, 0) + 1
    
    # Filter keywords that appear more than once
    repeated_keywords = {word: count for word, count in keyword_count.items() if count > 1}
    
    # Display the result
    print(repeated_keywords)



# input
query = "Buy mobile phone buy phone online"

print(f'Input: "{query}"')
print("Output:", end=" ")
# Calling the function
analyze_search_keywords(query)
```

### Problem 3: Sensor Data Validation

```python
def validate_sensor_readings(sensor_readings):

    # Find valid readings (even numbers) with their hour index
    valid_readings = []
    
    for hour, reading in enumerate(sensor_readings):
        if reading % 2 == 0:  # Check if reading is even (valid)
            valid_readings.append((hour, reading))
    
    # Display the result
    print(f"Valid Sensor Readings (Hour, Value): {valid_readings}")


# input
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

print(f"Input: {sensor_readings}")
validate_sensor_readings(sensor_readings)
```

### Problem 4: Email Domain Usage Analysis

```python
def analyze_email_domains(emails):

    # Count domain occurrences
    domain_count = {}
    
    for email in emails:
        # Extract domain (part after @)
        domain = email.split('@')[1]
        domain_count[domain] = domain_count.get(domain, 0) + 1
        
    
    # Calculate and display percentages
    total_emails = len(emails)
    
    for domain, count in domain_count.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")



# input
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

print(f"Input: {emails}")
print("Output:")
analyze_email_domains(emails)
```

### Problem 5: Sales Spike Detection

```python
def detect_sales_spikes(sales, threshold_percentage=30):

    # Calculate average sales
    average_sales = sum(sales) / len(sales)
    
    # Calculate threshold value
    threshold_value = average_sales + average_sales * (threshold_percentage / 100)
    print("Threshold Value: ", threshold_value)
    
    # Detect and display spikes
    for day, sale in enumerate(sales, 1):
        if sale > threshold_value:
            print(f"Day {day}: {sale}")


# input
sales = [1200, 1500, 900, 2200, 1400, 3000]

print(f"Input: {sales}")
print("Output:")
detect_sales_spikes(sales)
```

### Problem 6: Duplicate User ID Detection

```python
def detect_duplicate_ids(user_ids):
    
    # Count occurrences of each user ID
    id_count = {}
    for user_id in user_ids:
        id_count[user_id] = id_count.get(user_id, 0) + 1
    
    # Display only duplicates (count > 1)
    for user_id, count in id_count.items():
        if count > 1:
            print(f"{user_id} → {count} times")


# input
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

print(f"Input: {user_ids}")
print("Output:")
detect_duplicate_ids(user_ids)
```
