## Task 1: Conceptual Understanding

### 1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing (NLP), the words "Love" and "love" are treated as two different words because of the difference in uppercase and lowercase letters. Even though they have the same meaning, a computer sees them as separate tokens. This can lead to duplication and inconsistency in the data.

To solve this problem, we usually convert all text into lowercase or uppercase during preprocessing so that both "Love" and "love" are treated the same.

### 2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many common words like "is", "the", "and", etc., which do not add much meaning to the text. These words increase noise in the data and can affect the performance of the model.

Also, keeping unnecessary words makes the processing slower because the model has to handle more data than required.

### 3. Two real-world scenarios where removing stopwords can be harmful

Removing stopwords is not always useful and can sometimes change the meaning of a sentence.

One example is in sentiment analysis. Words like "not" are very important. For example, "I am not happy" becomes "I am happy" if "not" is removed, which gives the wrong meaning.

Another example is in question answering systems. Words like "what", "where", and "who" are necessary to understand the question properly. Removing them can make the sentence unclear.

### 4. Difference between stemming and lemmatization

Stemming and lemmatization are both techniques used to reduce words to their base form, but they work in different ways.

Stemming simply cuts off the end of a word, which may not always give a meaningful result. For example, "running" may become "runn".

Lemmatization, on the other hand, converts the word into its correct base form using proper language rules. For example, "running" becomes "run".

In short, stemming is faster but less accurate, while lemmatization is slower but gives more meaningful results.

## Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return []

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # 3. Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # 4. Remove numbers
    text = re.sub(r"\d+", "", text)

    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # 6. Remove special characters (keep only alphabets and spaces)
    text = re.sub(r"[^a-z\s]", "", text)

    # 7. Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # 8. Tokenization
    tokens = text.split()

    # 9. Remove short tokens (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens
```

```python
# Test example
sample_text = "WOW!! This is soooo GOOD!!! Visit https://innomatics.com now!!!"

tokens = preprocess_text(sample_text)
cleaned_sentence = " ".join(tokens)

print("Original:", sample_text)
print("Tokens:", tokens)
print("Cleaned Sentence:", cleaned_sentence)
```

## Task 3: Stress Testing

```python
# Test sentences (given in assignment)
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://innomatics.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Apply preprocessing and display results
for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 60)
```

## Task 4: Token Analytics

```python
def token_analysis(tokens):
    if not tokens:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_length, 2)


# Apply analysis on each sentence
for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    total, unique, avg = token_analysis(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg)
    print("-" * 60)
```

## Task 5: Frequency Analysis

```python
from collections import Counter

# Combine all tokens from all sentences
all_tokens = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Create frequency counter
word_freq = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, count in word_freq.most_common(10):
    print(word, ":", count)

print("\n" + "-"*50)

# Top 5 least frequent words
print("Top 5 Least Frequent Words:")
least_common = word_freq.most_common()[-5:]

for word, count in least_common:
    print(word, ":", count)
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)  # reuse Task 2 function
        
        all_tokens.extend(tokens)  # collect all tokens
        clean_sentences.append(" ".join(tokens))  # create cleaned sentence

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
# Run pipeline on test data
output = full_pipeline(test_sentences)

print("All Tokens:")
print(output["tokens"])

print("\nClean Sentences:")
for sentence in output["clean_sentences"]:
    print(sentence)
```
